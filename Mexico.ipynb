```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Mexico_City_Hostels.csv", index_col=[0])
df.head()
df_copy = df.copy()

df_copy
```

```python
new_df = df_copy.T.reset_index()
new_df
```

```python
df.shape
```

```python
df_test = df_copy.set_axis(['Security', 'Location', 'Staff', 'Atmosphere', 'Cleanliness', 'Value for Money', 'Facilities'], axis=0, copy=True)
```

```python
# df_copy.mean(axis=0)
```

```python
df_test.loc['avg']= df.mean()
df_test
```

```python
df_test
```

```python

avg_info = df_test.sort_values(by='avg', ascending=False)
figure = plt.figure(figsize=(10,6))
sns.barplot(y=avg_info.name, x=avg_info.ratings)
plt.xticks()
plt.xlabel('Ratings')
plt.ylabel('Hostel Name')
plt.title('Top 20 Hostels by Ratings')
plt.show()
```
