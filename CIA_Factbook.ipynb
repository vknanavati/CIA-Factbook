```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

URL = "https://www.cia.gov/the-world-factbook/countries/switzerland/travel-facts"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
soup.prettify()
```

```python
for topic in soup.find_all('h3'):
    print(topic.text)
    info = topic.find_next_sibling('p')
    print(info.text.strip())
    print()
    
```

```python
results_elements = soup.find_all("div", class_="col-lg-9 col-md-12 col-sm-12")
print(results_elements)
for result_element in results_elements:
    print(result_element.text.strip())
    print("")

# print(soup.h3) #only gives you the first tag

#find_all gives you all the tags
# h3_tag = soup.find_all('h3')

    
```

```python
app_elements = soup.find_all("h3", class_="mt30")
for app_element in app_elements:
    print(app_element.text.strip())
    print("")

```
